```python
# --- Train & Test: Imputation strategy ---
def checkNullValues(df):
    nulls = df.isnull().sum()
    nulls = nulls[nulls > 0]
    return nulls

# Electrical: mode (only 1 missing value)
most_common_elec = train_data['Electrical'].mode()[0]
train_data['Electrical'] = train_data['Electrical'].fillna(most_common_elec)
test_data['Electrical'] = test_data['Electrical'].fillna(most_common_elec)

# MasVnrArea: mode (MasVnrArea and MasVnrType usually go together)
most_common_masvnr = train_data['MasVnrArea'].mode()[0]
train_data['MasVnrArea'] = train_data['MasVnrArea'].fillna(most_common_masvnr)
test_data['MasVnrArea'] = test_data['MasVnrArea'].fillna(most_common_masvnr)

# MasVnrType: 'None'
train_data['MasVnrType'] = train_data['MasVnrType'].fillna('None')
test_data['MasVnrType'] = test_data['MasVnrType'].fillna('None')

# LotFrontage: median by Neighborhood
train_data['LotFrontage'] = train_data.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)
test_data['LotFrontage'] = test_data.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)

# Bsmt-related columns: 'None'
bsmt_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in bsmt_cols:
    train_data[col] = train_data[col].fillna('None')
    test_data[col] = test_data[col].fillna('None')

# Garage-related columns
train_data['GarageType'] = train_data['GarageType'].fillna("None")
test_data['GarageType'] = test_data['GarageType'].fillna("None")

train_data['GarageYrBlt'] = train_data['GarageYrBlt'].fillna(0)
test_data['GarageYrBlt'] = test_data['GarageYrBlt'].fillna(0)

train_data['GarageFinish'] = train_data['GarageFinish'].fillna("None")
test_data['GarageFinish'] = test_data['GarageFinish'].fillna("None")

train_data['GarageQual'] = train_data['GarageQual'].fillna("None")
test_data['GarageQual'] = test_data['GarageQual'].fillna("None")

train_data['GarageCond'] = train_data['GarageCond'].fillna("None")
test_data['GarageCond'] = test_data['GarageCond'].fillna("None")

# Others that mean "no feature" when missing
fill_none_cols = ['FireplaceQu', 'Fence', 'Alley', 'MiscFeature', 'PoolQC']
for col in fill_none_cols:
    train_data[col] = train_data[col].fillna("None")
    test_data[col] = test_data[col].fillna("None")

# MSZoning: Fill with mode
most_common_mszoning = train_data['MSZoning'].mode()[0]
test_data['MSZoning'] = test_data['MSZoning'].fillna(most_common_mszoning)

# Utilities: Fill with mode (most likely "AllPub")
most_common_utilities = train_data['Utilities'].mode()[0]
test_data['Utilities'] = test_data['Utilities'].fillna(most_common_utilities)

# Exterior1st and Exterior2nd: Fill with mode
test_data['Exterior1st'] = test_data['Exterior1st'].fillna(train_data['Exterior1st'].mode()[0])
test_data['Exterior2nd'] = test_data['Exterior2nd'].fillna(train_data['Exterior2nd'].mode()[0])

# Basement-related numerical features: Houses with no basement should have 0
for col in ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']:
    test_data[col] = test_data[col].fillna(0)

# BsmtFullBath and BsmtHalfBath: Houses with no basement should have 0 bathrooms
test_data['BsmtFullBath'] = test_data['BsmtFullBath'].fillna(0)
test_data['BsmtHalfBath'] = test_data['BsmtHalfBath'].fillna(0)

# KitchenQual: Fill with mode
test_data['KitchenQual'] = test_data['KitchenQual'].fillna(train_data['KitchenQual'].mode()[0])

# Functional: Fill with mode (most likely "Typ")
test_data['Functional'] = test_data['Functional'].fillna(train_data['Functional'].mode()[0])

# GarageCars and GarageArea: Houses with no garage should have 0
test_data['GarageCars'] = test_data['GarageCars'].fillna(0)
test_data['GarageArea'] = test_data['GarageArea'].fillna(0)

# SaleType: Fill with mode
test_data['SaleType'] = test_data['SaleType'].fillna(train_data['SaleType'].mode()[0])

print("Null values after imputation (test_data): ", checkNullValues(test_data))
print("Null values after imputation (train_data): ", checkNullValues(train_data))
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# --- Train & Test: Imputation strategy ---
def checkNullValues(df):
    nulls = df.isnull().sum()
    nulls = nulls[nulls > 0]
    return nulls

# Electrical: mode (only 1 missing value)
most_common_elec = train_data['Electrical'].mode()[0]
train_data['Electrical'] = train_data['Electrical'].fillna(most_common_elec)
test_data['Electrical'] = test_data['Electrical'].fillna(most_common_elec)

# MasVnrArea: mode (MasVnrArea and MasVnrType usually go together)
most_common_masvnr = train_data['MasVnrArea'].mode()[0]
train_data['MasVnrArea'] = train_data['MasVnrArea'].fillna(most_common_masvnr)
test_data['MasVnrArea'] = test_data['MasVnrArea'].fillna(most_common_masvnr)

# MasVnrType: 'None'
train_data['MasVnrType'] = train_data['MasVnrType'].fillna('None')
test_data['MasVnrType'] = test_data['MasVnrType'].fillna('None')

# LotFrontage: median by Neighborhood
train_data['LotFrontage'] = train_data.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)
test_data['LotFrontage'] = test_data.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)

# Bsmt-related columns: 'None'
bsmt_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in bsmt_cols:
    train_data[col] = train_data[col].fillna('None')
    test_data[col] = test_data[col].fillna('None')

# Garage-related columns
train_data['GarageType'] = train_data['GarageType'].fillna("None")
test_data['GarageType'] = test_data['GarageType'].fillna("None")

train_data['GarageYrBlt'] = train_data['GarageYrBlt'].fillna(0)
test_data['GarageYrBlt'] = test_data['GarageYrBlt'].fillna(0)

train_data['GarageFinish'] = train_data['GarageFinish'].fillna("None")
test_data['GarageFinish'] = test_data['GarageFinish'].fillna("None")

train_data['GarageQual'] = train_data['GarageQual'].fillna("None")
test_data['GarageQual'] = test_data['GarageQual'].fillna("None")

train_data['GarageCond'] = train_data['GarageCond'].fillna("None")
test_data['GarageCond'] = test_data['GarageCond'].fillna("None")

# Others that mean "no feature" when missing
fill_none_cols = ['FireplaceQu', 'Fence', 'Alley', 'MiscFeature', 'PoolQC']
for col in fill_none_cols:
    train_data[col] = train_data[col].fillna("None")
    test_data[col] = test_data[col].fillna("None")

# MSZoning: Fill with mode
most_common_mszoning = train_data['MSZoning'].mode()[0]
test_data['MSZoning'] = test_data['MSZoning'].fillna(most_common_mszoning)

# Utilities: Fill with mode (most likely "AllPub")
most_common_utilities = train_data['Utilities'].mode()[0]
test_data['Utilities'] = test_data['Utilities'].fillna(most_common_utilities)

# Exterior1st and Exterior2nd: Fill with mode
test_data['Exterior1st'] = test_data['Exterior1st'].fillna(train_data['Exterior1st'].mode()[0])
test_data['Exterior2nd'] = test_data['Exterior2nd'].fillna(train_data['Exterior2nd'].mode()[0])

# Basement-related numerical features: Houses with no basement should have 0
for col in ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']:
    test_data[col] = test_data[col].fillna(0)

# BsmtFullBath and BsmtHalfBath: Houses with no basement should have 0 bathrooms
test_data['BsmtFullBath'] = test_data['BsmtFullBath'].fillna(0)
test_data['BsmtHalfBath'] = test_data['BsmtHalfBath'].fillna(0)

# KitchenQual: Fill with mode
test_data['KitchenQual'] = test_data['KitchenQual'].fillna(train_data['KitchenQual'].mode()[0])

# Functional: Fill with mode (most likely "Typ")
test_data['Functional'] = test_data['Functional'].fillna(train_data['Functional'].mode()[0])

# GarageCars and GarageArea: Houses with no garage should have 0
test_data['GarageCars'] = test_data['GarageCars'].fillna(0)
test_data['GarageArea'] = test_data['GarageArea'].fillna(0)

# SaleType: Fill with mode
test_data['SaleType'] = test_data['SaleType'].fillna(train_data['SaleType'].mode()[0])

print("Null values after imputation (test_data): ", checkNullValues(test_data))
print("Null values after imputation (train_data): ", checkNullValues(train_data))

# CATEGORICAL TO NUMERICAL CONVERSION (train + test)

# 1. Define nominal columns (will use one-hot encoding)
nominal_cols = ['MSZoning', 'Alley', 'LotConfig', 'LandSlope', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType',
                'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',
                'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC',
                'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']

# Combine train and test for consistent one-hot encoding
all_data = pd.concat([train_data.drop('SalePrice', axis=1), test_data])
all_data_dummies = pd.get_dummies(all_data, columns=nominal_cols, drop_first=True)

# Split back to train and test
train_data_dummies = all_data_dummies.iloc[:len(train_data)]
test_data_dummies = all_data_dummies.iloc[len(train_data):]

# Add SalePrice back to train data
train_data_dummies['SalePrice'] = train_data['SalePrice']

# Update our variables
train_data = train_data_dummies
test_data = test_data_dummies

# 2. Ordinal mappings
# --- Street
street_map = {'Grvl': 0, 'Pave': 1}
train_data['Street'] = train_data['Street'].map(street_map)
test_data['Street'] = test_data['Street'].map(street_map)

# --- LotShape
lotshape_map = {'Reg': 3, 'IR1': 2, 'IR2': 1, 'IR3': 0}
train_data['LotShape'] = train_data['LotShape'].map(lotshape_map)
test_data['LotShape'] = test_data['LotShape'].map(lotshape_map)

# --- LandContour
landcontour_map = {'Low': 0, 'Bnk': 1, 'HLS': 1, 'Lvl': 2}
train_data['LandContour'] = train_data['LandContour'].map(landcontour_map)
test_data['LandContour'] = test_data['LandContour'].map(landcontour_map)

# --- Drop 'Utilities' (has very little variation)
train_data.drop('Utilities', axis=1, inplace=True)
test_data.drop('Utilities', axis=1, inplace=True)

# --- Neighborhood (ordinal via SalePrice)
neigh_order = train_data.groupby('Neighborhood')['SalePrice'].median().sort_values().index
neigh_map = {name: i for i, name in enumerate(neigh_order)}
train_data['Neighborhood'] = train_data['Neighborhood'].map(neigh_map)
test_data['Neighborhood'] = test_data['Neighborhood'].map(neigh_map)

# --- Condition1 (ordinal via SalePrice)
cond1_order = train_data.groupby('Condition1')['SalePrice'].median().sort_values().index
cond1_map = {name: i for i, name in enumerate(cond1_order)}
train_data['Condition1'] = train_data['Condition1'].map(cond1_map)
test_data['Condition1'] = test_data['Condition1'].map(cond1_map)

# --- Drop 'Condition2' 
if 'Condition2' in train_data.columns:
    train_data.drop('Condition2', axis=1, inplace=True)
if 'Condition2' in test_data.columns:
    test_data.drop('Condition2', axis=1, inplace=True)

# --- BldgType → binary: IsSingleFamily
if 'BldgType' in train_data.columns:
    train_data['IsSingleFamily'] = (train_data['BldgType'] == '1Fam').astype(int)
    test_data['IsSingleFamily'] = (test_data['BldgType'] == '1Fam').astype(int)
    train_data.drop(columns=['BldgType'], inplace=True)
    test_data.drop(columns=['BldgType'], inplace=True)

# --- HouseStyle (ordinal via SalePrice)
if 'HouseStyle' in train_data.columns:
    housestyle_order = train_data.groupby('HouseStyle')['SalePrice'].median().sort_values().index
    housestyle_map = {name: i for i, name in enumerate(housestyle_order)}
    train_data['HouseStyle'] = train_data['HouseStyle'].map(housestyle_map)
    test_data['HouseStyle'] = test_data['HouseStyle'].map(housestyle_map)

# --- RoofStyle (ordinal via SalePrice)
if 'RoofStyle' in train_data.columns:
    roofstyle_order = train_data.groupby('RoofStyle')['SalePrice'].median().sort_values().index
    roofstyle_map = {name: i for i, name in enumerate(roofstyle_order)}
    train_data['RoofStyle'] = train_data['RoofStyle'].map(roofstyle_map)
    test_data['RoofStyle'] = test_data['RoofStyle'].map(roofstyle_map)

# Handle any remaining non-numeric columns
X = train_data.drop(columns=['SalePrice'])
remaining_non_numeric = X.select_dtypes(exclude='number').columns.tolist()

if remaining_non_numeric:
    print(f"Remaining non-numeric columns: {remaining_non_numeric}")
    
    for col in remaining_non_numeric:
        # Check if the column exists in both dataframes
        if col in train_data.columns and col in test_data.columns:
            # For columns with few unique values, use one-hot encoding
            if train_data[col].nunique() < 10:
                print(f"One-hot encoding: {col}")
                train_dummies = pd.get_dummies(train_data[col], prefix=col, drop_first=True)
                test_dummies = pd.get_dummies(test_data[col], prefix=col, drop_first=True)
                
                # Add the dummy columns to the dataframes
                train_data = pd.concat([train_data, train_dummies], axis=1)
                test_data = pd.concat([test_data, test_dummies], axis=1)
                
                # Remove the original column
                train_data.drop(columns=[col], inplace=True)
                test_data.drop(columns=[col], inplace=True)
            else:
                # For columns with many unique values, use label encoding
                print(f"Label encoding: {col}")
                le = LabelEncoder()
                # Combine values from both train and test to fit encoder
                combined_values = pd.concat([train_data[col], test_data[col]]).unique()
                le.fit(combined_values)
                
                # Transform the columns
                train_data[col] = le.transform(train_data[col])
                test_data[col] = le.transform(test_data[col])

# Final alignment and check
# Extract features from train data (excluding SalePrice)
X = train_data.drop(columns=['SalePrice'])
y = train_data['SalePrice']

# Check for column differences between train and test
train_cols = set(X.columns)
test_cols = set(test_data.columns)

print(f"Columns only in train: {train_cols - test_cols}")
print(f"Columns only in test: {test_cols - train_cols}")

# Add missing columns to each dataset
for col in train_cols - test_cols:
    test_data[col] = 0  # Add missing columns to test with default value 0

for col in test_cols - train_cols:
    X[col] = 0  # Add missing columns to train with default value 0
    train_data[col] = 0  # Also add to train_data

# Ensure same column order
X_cols = sorted(X.columns)
test_cols = sorted(test_data.columns)

# Reindex both datasets
X = X[X_cols]
test_data = test_data[test_cols]
train_data_cols = X_cols + ['SalePrice']
train_data = train_data[train_data_cols]

# Final check
assert X.select_dtypes(exclude='number').empty, "❌ Non-numeric columns in train features!"
assert test_data.select_dtypes(exclude='number').empty, "❌ Non-numeric columns in test features!"
assert list(X.columns) == list(test_data.columns), "❌ Train and test column mismatch!"

print("✅ Data is clean, numeric, and aligned. Ready for modeling!")
```
